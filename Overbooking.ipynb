```python
# plot showing proportion between cancelled and not cancelled bookings
cancel_counts = df['is_canceled'].value_counts()
labels = ['Not Canceled', 'Canceled']
colors = ['#66b3ff', '#ff9999']

plt.pie(cancel_counts, labels = labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=(0,0.1))
plt.title('Booking cancelation')
plt.axis('equal')
plt.show()
```

```python
#Visualisation to show the cancellation rate by customer type
plt.figure(figsize=(8, 5))
sns.barplot(x='customer_type', y='is_canceled', data=df, estimator=lambda x: sum(x)/len(x))
plt.title('Cancellation Rate by Customer Type')
plt.ylabel('Cancellation Rate')
plt.xlabel('Customer Type')
plt.ylim(0, 1)
plt.show()
```
